```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
lego_data = pd.read_csv('./DATA/lego_sets.csv')
lego_data = lego_data.astype({"list_price": int})
```

# Nous allons étuider des donénes sur des sets de lego. 

```python
lego_data.describe()
```

```python
lego_data.head()
```

## Grouper et identifier les themes: 

Nous voyons donc que les sets de lego sont groupé principalement par theme. Nous allons donc étudier les disparités entre ces groupes. 

```python
mean_theme = lego_data.groupby('theme_name').mean()
mean_theme
```

```python
mean_theme.info()
```

## Résultat du tri 

Il y a donc 40 themes différents 

Remarquon l'effet du nombre de pièce sur le prix du set. 

```python
sns.relplot(x="list_price",y="piece_count",data=mean_theme, height=10, hue='theme_name')
```

```python
sns.relplot(x="list_price",y="piece_count",data=mean_theme.query('piece_count > 100 and piece_count < 500'), height=10, hue='theme_name')
```

On vois que le prix a tendance à augmenter avec le nombre de piece dans la boite, ainsi que par l'influence du theme. 
En choissisant des themes parmis ceux du set nous allons pouvoir comparé l'influence du theme sur le prix. 

## Filtrage par themes

```python
lego_filtre_theme = lego_data[(lego_data['theme_name'] == 'Star Wars™') | (lego_data['theme_name'] == 'Classic') | (lego_data['theme_name'] == 'Marvel Super Heroes') | (lego_data['theme_name'] == 'Minecraft™' ) | (lego_data['theme_name'] == 'NINJAGO®') | (lego_data['theme_name'] == 'City') ]	
lego_filtre_theme.describe()
```

```python
sns.distplot(lego_filtre_theme.star_rating, color='g')
```

```python
sns.distplot(lego_data.star_rating, color='g')
```

## Analyse de la notation 
On vous que la courbe de notation est globalement similaire

```python
sns.catplot(x='theme_name', y='list_price', kind='box', data=lego_filtre_theme)
```

Sans filtre, on peux voire qu'il y a des données abérantes. il s'agit des boites contenants beaucoup de pieces, surement issu de collections spéciale. 

```python
sns.catplot(x='theme_name', y='list_price', kind='box', data=lego_filtre_theme.query('piece_count > 270.000 and piece_count < 500'), height=10)
```

En filtrant sur un nombre de piece compris entre la mediane et le dernier quartile, on obtient des résultats sans donénes abérantes. 
On vois alors que les boites star wars et minecraft sont en moyenne plus cher, surtout si on les compare aux sets "classique" de la gamme lego.  

les écarts types quant à eux sont plus réduit sur les sets classiques, on peut en déduires que les sets classiques ont a peu prés un nombre de pieces égales entres les sets. 

```python
sns.catplot(x='theme_name', y='piece_count', kind='box', data=lego_filtre_theme.query('piece_count > 270.000 and piece_count < 500'), height=10)
```

## Comptage de la répartition du volume de boite par set

```python
plt.subplots(figsize=(30,10))
sns.countplot(data=lego_filtre_theme, x='theme_name')
```

La franchise star wars est la plus édité, il existe donc plus de set de lego star wars, que de lego classique. 
*On en déduis que lego vends plus de sets avec la trademark star wars, plutot que ses sets classiques 

```python
corr = lego_filtre_theme.corr()
corr
```

```python
f, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corr, annot=True, ax =ax)
plt.savefig('heatmap_Lego_theme.png')
```
