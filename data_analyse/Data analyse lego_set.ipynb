```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
lego_data = pd.read_csv('./DATA/lego_sets.csv')
```

```python

lego_filtre = lego_data[(lego_data['star_rating']>4)] 
lego_filtre.head()
```

```python
lego_filtre_theme = lego_data[(lego_data['theme_name'] == 'Star Wars™') | (lego_data['theme_name'] == 'Classic') | (lego_data['theme_name'] == 'Marvel Super Heroes') | (lego_data['theme_name'] == 'Minecraft™' ) | (lego_data['theme_name'] == 'Friends') ]
lego_filtre_theme.head()
```

```python
plt.subplots(figsize=(30,10))
sns.countplot(x='theme_name', data=lego_data)
```

```python
filtre_theme_used = lego_data.groupby("theme_name").sum()
filtre_theme_used.head()
```
